```python
# Importar bibliotecas
import pandas as pd
from pathlib import Path

# Definir caminho do arquivo CSV
RAW_PATH = Path("../data/raw/creditcard.csv")

# Ler o CSV
df = pd.read_csv(RAW_PATH)

# Mostrar as 5 primeiras linhas
df.head()
```

```python
# Ver informações gerais do dataset
df.info()
```

```python
# Estatísticas descritivas das colunas numéricas
df.describe()
```

```python
# Verificar se há valores nulos
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contagem de classes
sns.countplot(x='Class', data=df)
plt.title('Distribuição de Transações Normais e Fraudes')
plt.show()

# Percentual de fraudes
fraude_percent = df['Class'].value_counts(normalize=True) * 100
print(fraude_percent)
```
